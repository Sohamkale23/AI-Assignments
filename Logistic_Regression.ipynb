```python
# Select only numeric columns for outlier removal
numeric_data = data.select_dtypes(include=[np.number])

# Compute IQR
q1 = numeric_data.quantile(0.25)
q3 = numeric_data.quantile(0.75)
iqr = q3 - q1

# Filter out outliers only for numeric columns
filtered_data = numeric_data[~((numeric_data < (q1 - 1.5 * iqr)) | (numeric_data > (q3 + 1.5 * iqr))).any(axis=1)]

# Keep non-numeric columns and merge with filtered numeric data
data = data.loc[filtered_data.index]

print("Outliers removed. New shape:", data.shape)
```
